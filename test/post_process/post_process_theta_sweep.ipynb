```python
import pandas as pd
df = pd.read_csv("../theta_sweep.csv")
display(df)
```

```python
df_rise = df[df["PULSE_POS"]=="FALL"]
df_rise.plot(x="THETA_PS", y="HW_OUT", marker = 'o', title='Falling Transition w/ Tclk=40ns')
df_fall = df[df["PULSE_POS"]=="RISE"]
df_fall.plot(x="THETA_PS", y="HW_OUT_INV", marker = 'o', title='Rising Transition  w/ Tclk=40ns')
```

```python
df_fall = df[(df["PULSE_POS"]=="FALL") & (df["HW_SAMPLES"] != "x")]
df_fall["THETA_PS"] = pd.to_numeric(df_fall["THETA_PS"])
df_fall["HW_SAMPLES"] = pd.to_numeric(df_fall["HW_SAMPLES"])
df_fall.plot(x="THETA_PS", y="HW_SAMPLES", marker = 'o', title='Falling Transition  w/ Tclk=20ns')
df_rise = df[(df["PULSE_POS"]=="RISE") & (df["HW_SAMPLES"]!= "x")]
df_rise["THETA_PS"] = pd.to_numeric(df_rise["THETA_PS"])
df_rise["HW_SAMPLES"] = pd.to_numeric(df_rise["HW_SAMPLES"])
df_rise.plot(x="THETA_PS", y="HW_SAMPLES", marker = 'o', title='Rising Transition  w/ Tclk=20ns')
```
